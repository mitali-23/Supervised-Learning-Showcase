```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model

import warnings

warnings.filterwarnings('ignore')
```

```python
data={
    'area':[1000,1500,1200,1800,2000,2500,1600,2200,1900,2400],
    'price':[200000,300000,250000,350000,400000,500000,320000,450000,380000,480000]
}
```

```python
df=pd.DataFrame(data)
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
x=df[['area']]
x
```

```python
y=df[['price']]
y
```

```python
reg=linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[2500]])
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the entire dataset
y_pred = reg.predict(df[['area']])

# Calculate MSE
mse = mean_squared_error(df.price, y_pred)
print(f"Mean Squared Error: {mse}")

# Calculate R-squared
r2 = r2_score(df['price'], y_pred)
print(f"R-squared: {r2}")
```
